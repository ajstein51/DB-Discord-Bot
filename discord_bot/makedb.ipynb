```python
# create the database file if needed
def create_connection(db_file):
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        return conn
    except Error as e:
        print(e)

    return conn
# end of fun

# make the db if needed, all in 1 database may need to break up in the future. Also makes the connection
def start_db():
    dbcon = create_connection('game.db')
    if dbcon is not None and dbcon != 0:
        df.to_sql('game.db', dbcon, index=True)
    else:
        print("error in making table")
    
    return dbcon
# end of fun

# # will create a table if it doesnt already exist
dbconnection = start_db()
```
